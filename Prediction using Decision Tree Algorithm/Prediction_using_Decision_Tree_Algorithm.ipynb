```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
# Load the dataset from a CSV file
data = pd.read_csv('Iris.csv')
```

```python
# Extract features and labels
X = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = data['Species']
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create a Decision Tree classifier
clf = DecisionTreeClassifier()
```

```python
# Train the classifier on the training data
clf.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = clf.predict(X_test)
```

```python
# Evaluate the classifier
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Visualize the Decision Tree
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=X.columns, class_names=y.unique(), filled=True)
plt.show()
```

```python
# New data to predict
new_data = pd.DataFrame([[5.1, 3.5, 1.4, 0.2], [6.7, 3.0, 5.2, 2.3]], columns=X.columns)
```

```python
# Make predictions on new data
new_predictions = clf.predict(new_data)
```

```python
# Print the predicted classes
for data, predicted_class in zip(new_data.values, new_predictions):
    print("Data:", data)
    print("Predicted Class:", predicted_class)
    print()
```
