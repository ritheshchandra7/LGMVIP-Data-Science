# Image to Pencil Sketch with Python

```python
%matplotlib inline
```

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import display,clear_output
```

```python
image=cv2.imread('Zebra.jpg')
```

```python
display(Image.fromarray(image))
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

```python
inverted_image = 255 - gray_image
display(Image.fromarray(inverted_image))
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
invert_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, invert_blurred, scale=256.0)
display(Image.fromarray(pencil_sketch))
```

```python
plt.figure(figsize=(18, 6))
```

```python
# Display the color image on the left subplot
plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title('Color Image')
plt.axis('off')

# Display the grayscale image in the middle subplot
plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Display the pencil sketch image on the right subplot
plt.subplot(1, 3, 3)
plt.imshow(pencil_sketch, cmap='gray')
plt.title('Pencil Sketch')
plt.axis('off')

# Show the figure with all three images
plt.show()
```
